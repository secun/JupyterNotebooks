```python
from matplotlib import pylab as plt
import math
import numpy as np
```

```python
def values_inic(ang,v0):
    """Pensado para angulos en grados que se pasan a radianes. 
    La velocidad inicial tiene que estar en m/s. 
    La salida son tres valores: vx, vy y t_tot"""
    
    
    angr = math.radians(ang)
    
    u = v0 * math.sin(angr)
    v = v0 * math.cos(angr)
    t_tot = (2*v)/g
    
    return (u,v,t_tot)
```

```python
u,v,tt=values_inic(30,20)
print(u,v,tt)
```

```python
def positions(u,v,tt):
    """Calculamos las posiciones x e y para el tiempo total.
    Elegimos un valor de 2000 puntos en linspace.
    
    Valores de entrada: vx, vy y tiempo total.
    
    Valores de salida: listas de posiciones x e posiciones y """
    x = []
    y = []
    intervalo = np.linspace(0,tt, 2000)
    
    for t in intervalo:
        x.append(u*t) 
        y.append(v*t - 0.5*g*t**2)
        
    return x,y
```

```python
xx,yy = positions(u,v,tt)
xx
```

```python
def draw_tiro(t_tot):
    plt.plot(xx,yy)
    plt.xlabel('coordenada x')
    plt.ylabel('coordenada y')
    plt.title('Tiro parabólico')
    plt.text(20,10,'Tiempo total del tiro {0:.2f} s'.format(t_tot))
    plt.show()
```

```python
ang = 20
v0 = 50
g = 9.8

uu,vv,tt = values_inic(ang,v0)
xx,yy = positions(uu,vv,tt)
draw_tiro(tt)
```
